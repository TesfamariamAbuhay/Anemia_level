```python
#Basic libraries
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
from collections import Counter
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
# library used for over sampling the data
from imblearn.over_sampling import RandomOverSampler
```

```python
dataset = pd.read_csv('C:/Users/Bel_Z_Man/Desktop/data_set.csv',sep=',',index_col='id')
```

```python
dataset.shape
```

```python
X=dataset.drop('v457',axis=1)
y=dataset.v457
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
```

```python
X_train.shape,y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,roc_auc_score,precision_score,recall_score,f1_score
from catboost import CatBoostClassifier
model = CatBoostClassifier(depth=10,iterations=300,l2_leaf_reg=1,learning_rate=0.15)
model.fit(X_train,y_train,verbose=False, plot=True)
pred = model.predict(X_test)
print(classification_report(y_test,pred))
```

```python
accuracy=accuracy_score(y_test,pred)
accuracy
```

```python
print("",confusion_matrix(y_test,pred))
```

```python
precision=precision_score(y_test, pred, pos_label='positive',average='macro')
precision
```

```python
recall=recall_score(y_test, pred,pos_label='positive',average='macro')
recall
```

```python
from sklearn import model_selection
kfold = model_selection.KFold(n_splits=10)
results = model_selection.cross_val_score(model, X,y.values.ravel(),scoring='accuracy', cv=kfold)
```

```python
corss_val=results.mean()
corss_val
```

```python
f1_score=f1_score(y_test, pred,pos_label='positive',average='macro')
f1_score
```

```python
ROC_with_cat_ovo=roc_auc_score(y, model.predict_proba(X), multi_class='ovo')
ROC_with_cat_ovo
```

```python
ROC_with_cat_ovr=roc_auc_score(y, model.predict_proba(X), multi_class='ovr')
ROC_with_cat_ovr
```

```python
cat = pd.DataFrame({'accuracy':accuracy,
                   'precision':precision,'recall':recall,'corss_val':corss_val,'f1_score':f1_score,'ROC_with_cat_ovo':ROC_with_cat_ovo,'ROC_with_cat_ovr':ROC_with_cat_ovr},
       columns=['accuracy','precision','recall','corss_val','f1_score','ROC_with_cat_ovo','ROC_with_cat_ovr'],index=["performance"])
cat
```
